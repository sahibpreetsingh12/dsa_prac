1. **x** = list of corrdinates

3. **width** = single value by default width is 0.8


By default the width of each bar in `plt.bar()` is `0.8`

````python
import numpy as np
course_list = ['CSE','ME','EE','CE','Mining']
barIndexes_alongsidexaxis = np.arange(len(course_list))
print(barIndexes_alongsidexaxis)
girl_count_barHights = [150,100,50,20,8]
boy_count_barHights = [200,150,100,56,27]
````

````python
# iterating over whole dataset (each value in each row)
curr_index = 0
for x in np.nditer(dataset):
    print(x, curr_index)
    curr_index += 1
````
